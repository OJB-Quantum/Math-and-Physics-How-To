## Motivating the Schrödinger Equation from Classical Energy and Wave Mechanics

The Schrödinger equation is best introduced as a physically motivated postulate whose form is constrained by classical mechanics, de Broglie wave relations, and the requirement that quantum states evolve linearly and consistently in time. Although it is often presented through a derivation-like sequence, the equation is ultimately a postulate of nonrelativistic quantum mechanics whose validity is confirmed by experiment.

---

### Step 1: Begin with the Classical Energy Relation

For a nonrelativistic free particle with mass $m$, momentum $p$, and no external potential energy, the classical Hamiltonian is

$$E = \frac{p^2}{2m}.$$

This expression gives the total mechanical energy of a free particle. In quantum mechanics, the same energy-momentum structure is retained, although the quantities $E$ and $p$ are represented by operators acting on a wavefunction.

---

### Step 2: Introduce the de Broglie and Planck-Einstein Relations

A plane wave can be written as

$$\psi(x,t) = A e^{i(kx-\omega t)},$$

where $k$ is the wave number and $\omega$ is the angular frequency. The de Broglie and Planck-Einstein relations connect these wave quantities to particle quantities through

$$p = \hbar k,$$

$$E = \hbar \omega.$$

Here, $\hbar$ is the reduced Planck constant, $k = 2\pi/\lambda$, and $\omega = 2\pi f$. These relations provide the bridge between classical particle variables and wave-mechanical variables.

---

### Step 3: Identify the Quantum Operators

When the plane wave is differentiated with respect to position, one obtains

$$\frac{\partial \psi}{\partial x} = ik\psi.$$

Since $p = \hbar k$, momentum is represented in the position basis by the operator

$$\hat{p} = -i\hbar \frac{\partial}{\partial x}.$$

The plane wave is therefore an eigenfunction of the momentum operator:

$$\hat{p}\psi = p\psi.$$

Similarly, differentiating the plane wave with respect to time gives

$$\frac{\partial \psi}{\partial t} = -i\omega \psi.$$

Since $E = \hbar \omega$, energy is represented by the operator

$$\hat{E} = i\hbar \frac{\partial}{\partial t}.$$

For a plane wave with definite energy, this gives

$$\hat{E}\psi = E\psi.$$

These operators are defined by how they act on the wavefunction $\psi(x,t)$.

---

### Step 4: Apply the Operators to the Free-Particle Energy Relation

Starting from the classical relation

$$E = \frac{p^2}{2m},$$

we replace $E$ and $p$ by the corresponding quantum operators:

$$\hat{E}\psi = \frac{\hat{p}^2}{2m}\psi.$$

Using

$$\hat{E} = i\hbar \frac{\partial}{\partial t}$$

and

$$\hat{p} = -i\hbar \frac{\partial}{\partial x},$$

we obtain

$$i\hbar \frac{\partial \psi}{\partial t} = \frac{1}{2m} \left( -i\hbar \frac{\partial}{\partial x} \right)^2 \psi.$$

Since

$$\left( -i\hbar \frac{\partial}{\partial x} \right)^2 = -\hbar^2 \frac{\partial^2}{\partial x^2},$$

the free-particle time-dependent Schrödinger equation in one spatial dimension becomes

$$i\hbar \frac{\partial \psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \psi}{\partial x^2}.$$

---

### Step 5: Include a Potential Energy Term

For a particle moving in a real-valued scalar potential $V(x,t)$, the Hamiltonian operator becomes

$$\hat{H} = \frac{\hat{p}^2}{2m} + V(x,t).$$

Here, $V(x,t)$ acts as a multiplication operator on the wavefunction. Therefore,

$$i\hbar \frac{\partial \psi}{\partial t} = \left[ -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + V(x,t) \right]\psi.$$

In three spatial dimensions, the second derivative is replaced by the Laplacian operator $\nabla^2$, giving

$$i\hbar \frac{\partial \psi}{\partial t} = \left[ -\frac{\hbar^2}{2m}\nabla^2 + V(\mathbf{r},t) \right]\psi.$$

This is the time-dependent Schrödinger equation for a nonrelativistic particle in a real-valued scalar potential.

When the potential is time independent, the Hamiltonian is also time independent, and stationary-state solutions can be introduced. When the potential depends explicitly on time, the Hamiltonian still generates time evolution, although energy is generally not conserved.

For real-valued potentials and appropriate boundary conditions, the Hamiltonian is self-adjoint, which ensures norm-preserving quantum time evolution.

---

### Step 6: Compare with the Heat Equation

The heat equation in one spatial dimension is

$$\frac{\partial u}{\partial t} = D\frac{\partial^2 u}{\partial x^2},$$

where $D$ is a real diffusion coefficient. For the free-particle Schrödinger equation,

$$i\hbar \frac{\partial \psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2 \psi}{\partial x^2},$$

division by $i\hbar$ gives

$$\frac{\partial \psi}{\partial t} = \frac{i\hbar}{2m} \frac{\partial^2 \psi}{\partial x^2}.$$

The resulting equation is diffusion-like in form because it contains an effective diffusion coefficient $i\hbar/(2m)$. For a real-valued potential and boundary conditions that make the Hamiltonian self-adjoint, this imaginary coefficient generates unitary, oscillatory evolution with conserved norm instead of dissipative diffusion.

A formal Wick rotation, using an imaginary-time variable such as $\tau = it$, transforms the Schrödinger equation into an equation resembling diffusion:

$$\frac{\partial \psi}{\partial \tau} = -\frac{1}{\hbar}\hat{H}\psi.$$

For a free particle, this becomes

$$\frac{\partial \psi}{\partial \tau} = \frac{\hbar}{2m} \nabla^2 \psi.$$

This imaginary-time form is useful in statistical mechanics, path integrals, ground-state projection methods, and diffusion Monte Carlo.

---

### Final Summary

The Schrödinger equation is motivated by combining the classical Hamiltonian with the wave relations $p = \hbar k$ and $E = \hbar\omega$. In the position representation, the substitutions

$$\hat{p} = -i\hbar \frac{\partial}{\partial x}$$

and

$$\hat{E} = i\hbar \frac{\partial}{\partial t}$$

lead naturally to

$$i\hbar \frac{\partial \psi}{\partial t} = \hat{H}\psi.$$

For a nonrelativistic particle in a real-valued scalar potential, the Hamiltonian is

$$\hat{H} = -\frac{\hbar^2}{2m}\nabla^2 + V(\mathbf{r},t).$$

The heat-equation analogy is useful because both equations couple a first-order derivative in time to a second-order derivative in space. The essential distinction is that the Schrödinger equation contains the imaginary unit $i$, so a self-adjoint Hamiltonian generates unitary, norm-preserving wave evolution, whereas the heat equation generates real, dissipative diffusion.
